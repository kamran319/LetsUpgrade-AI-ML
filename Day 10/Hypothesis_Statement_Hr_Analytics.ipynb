# Hypothesis Conclusion from the above Data

1. All the features are postive except Education, conclusion draw based skewness(mean<median)
2. Age, DistanceFromHome, Education, EmployeeID and PercentSalaryHike are Platykurtic(negatively kurt as there peakness is low)
3. EmployeeCount and StandardHours are mesokurtic as there peakness is zero
4. Remaining all other features are leptokurtic as there peakness is positively kurt
5. From Age and Total_work_experience features we can conclude that as the age and Total_work_experience of the employee increases the salary too increases 
6. Features like MonthlyIncome, YearsAtCompany, YearsSinceLastPromotion,YearsWithCurrManager are rightly positively skewed, while plotting box plot
7. Features like Age, DistanceFromHome, Education, JobLevel, PercentSalaryHike has no outliers
8. StockOptionLevel feature has only one outlier that is rightly postive skewed
9. TrainingTimesLastYear feature has few outliers that is rightly postive and as well as leftly negative skewed
10. From interquartile range we can conclude that Age has 13years of IQR, MonthlyIncome is 54690 IQR, YearsSinceLastPromotion has 3years of IQR and PercentSalaryHike is 6%.
11. As per std deviation the monthly income can be concluded as consistent(47068.888559)
