# Statistical Test

## 1. Mannwhitenyu Test: It is used to compare to independent sample

```python
from scipy.stats import mannwhitneyu
```

```python
stats,p=mannwhitneyu(data_Yes.DistanceFromHome,data_No.DistanceFromHome)
print(stats,p)
```

___As p value is less than 0.05, thus we can conclude that alternative hypothesis is rejected i.e there is a no significant difference of the Distance from home between Attrition Yes and Attrition No___ 

```python
stats,p=mannwhitneyu(data_Yes.MonthlyIncome,data_No.MonthlyIncome)
print(stats,p)
```

___As p value is greater than 0.05, thus we can canclude that alternative hypothesis is accepted i.e there is a significant difference of the Monthly Income between Attrition Yes and Attrition No___

```python
stats,p=mannwhitneyu(data_Yes.TotalWorkingYears,data_No.TotalWorkingYears)
print(stats,p)
```

___As p value is less than 0.05, thus we can conclude that alternative hypothesis is rejected i.e there is a no significant difference TotalWorkingYears between Attrition Yes and Attrition No___ 

```python
stats,p=mannwhitneyu(data_Yes.YearsAtCompany,data_No.YearsAtCompany)
print(stats,p)
```

___As p value is less than 0.05, thus we can conclude that alternative hypothesis is rejected i.e there is a no significant difference of YeasAtCompany between Attrition Yes and Attrition No___ 

## Chi Square T Test : It is used to check the dependency of the categorical variables 

```python
from scipy.stats import chi2_contingency
```

```python
chitable=pd.crosstab(hr.BusinessTravel,hr.Over18)
```

```python
stats,p,dof,expected=chi2_contingency(chitable)
print(stats,p)
```

___As p value is greater than 0.05, thus we can conclude that alternative hypothesis is acseptedi.e there is a dependance between BusinessTravel and Over18___

## 4.One way anova: It is used when dependent variable is continuos and indepent variable is categorical

```python
from sklearn import preprocessing 
Label_encoder=preprocessing.LabelBinarizer()
```

```python
hr['Attrition'] = Label_encoder.fit_transform(hr['Attrition']) 
```

```python
hr.head()
```

```python
import statsmodels.api as sm
```

```python
from statsmodels.formula.api import ols
```

```python
models=ols("Attrition~C(Gender)",hr).fit()
```

```python
oneway=sm.stats.anova_lm(models,typ=2)
print(oneway)
```

___we can conclude that here between variation is 0.195918 and within variation is 596.1734 , here P value is less than 0.05 hence we can reject the null hypothesis i.e there is no signifance differenace between attrition and gender___

```python
models=ols("Attrition~C(BusinessTravel)",hr).fit()
```

```python
oneway=sm.stats.anova_lm(models,typ=2)
print(oneway)
```
