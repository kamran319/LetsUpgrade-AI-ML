## Statistical Test

## Chi Square

```python
from scipy.stats import chi2_contingency
```

```python
chitable=pd.crosstab(hr.BusinessTravel,hr.Over18)
```

```python
stats,p,dof,expected=chi2_contingency(chitable)
print(stats,p)
```

___As p value is less than 0.05, thus we can conclude that alternative hypothesis is rejected i.e there is a no dependance between Departement and EducationField___

```python
chitable=pd.crosstab(hr.Gender,hr.JobLevel)
```

```python
stats,p,dof,expected=chi2_contingency(chitable)
print(stats,p)
```

___As p value is less than 0.05, thus we can conclude that alternative hypothesis is rejected i.e there is a no dependance between Gender and Job role___ 

## One way Anova

```python
import statsmodels.api as sm
```

```python
from statsmodels.formula.api import ols
```

```python
models=ols("Attrition~C(Gender)",hr).fit()
```

```python
oneway=sm.stats.anova_lm(models,typ=2)
print(oneway)
```

___we can conclude that here between variation is 0.195918 and within variation is 596.1734 , here P value is less than 0.05 hence we can reject the null hypothesis i.e there is no signifance differenace between attrition and gender___
